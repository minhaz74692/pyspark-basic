```python
df = spark.read.format("csv").option("header",True).option("inferSchema", True).load("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
df.limit(5).display()
```

### APPEND MODE

```python
df.write.format("csv")\
    .option("header", "true") \
    .mode("append")\
    .save("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
```

### OVERWRITE MODE

```python
df.write.format("csv") \
    .option("header", "true") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
```

```python
df = spark.read.format("csv").option("header",True).option("inferSchema", True).load("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
df.count(*)
```

### ERROR MODE

```python
df.write.format("csv")\
    .mode("error")\
    .save("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
```

```python
df.write.format("csv")\
    .mode("ignore")\
    .save("/Volumes/workspace/default/volume1/BigMartSalesUpdated.csv")
```
