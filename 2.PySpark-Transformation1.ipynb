## PySpark Transformation 1

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv").option('inferSchema', True).option('header', True).load('/Volumes/workspace/default/volume1/BigMart Sales.csv')
```

```python
display(df)
```

## SELECT

```python
df.select('Item_Identifier', 'Item_Weight').display()
```

### SELECT by Col

```python
df.select(col('Item_Identifier'), col('Item_Weight')).display() 
```

##ALIAS

```python
df.select(col("Item_Identifier").alias("item_id"), col('Item_Weight').alias('item_weight')).display()
```

```python
df.display()
```

## FILTER, WHERE

### Scenario 1

```python
df.filter(col('Item_Fat_Content') == "Regular").display()
```

## Scenario 2

```python
df.filter((col("Item_Type") == "Soft Drinks") & (col("Item_Weight") <= 10)).display()
```

## Scenario 3

```python
df.filter((col("Outlet_Size").isNull()) & (col("Outlet_Location_Type").isin("Tier 1", "Tier 2"))).display ()
```

## WithColumnRenamed

```python
df.withColumnRenamed("Item_Weight", "Item_wt").display()
```

## withColumn

```python
df.withColumn("item_price", col("Item_Weight") * col("Item_MRP")).select("item_price").display()
```

```python
df.withColumn("flag", 
              when(col("Outlet_Establishment_Year") > 1999, "NEW")
              .otherwise("OLD")
).display()
```

## Modify A existign Column 

```python
df.withColumn("Item_Fat_content", regexp_replace(col("Item_Fat_Content"), "Regular", "Reg"))\
    .withColumn("Item_Fat_Content", regexp_replace(col("Item_Fat_Content"), "Low Fat", "LF"))\
    .display()
```

```python
df.withColumn(
    "Item_Fat_Content",
    when(col("Item_Fat_Content") == "Regular", "Reg")
    .when(col("Item_Fat_Content") == "Low Fat", "LF")
    .otherwise(col("Item_Fat_Content"))
).display()
```

## CASTING DataType

```python
df = df.withColumn("Item_Weight", col("Item_Weight").cast(StringType()))
```

```python
df.printSchema()
```

## Sort/Order By

```python
df.sort(col("Item_Weight").desc()).display()
```

```python
df.sort(col("Item_Visibility").asc()).display()
```

#Sorting in multiple columns 

```python
df.sort(["Item_Weight", "Item_Visibility"], ascending = [False, True]).limit(5).display()
```
