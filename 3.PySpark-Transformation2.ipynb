# UNION and UNION BY NAME

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
data1 = [
            ("1", "kad"),
            ("2", "sid")
        ]
schema1 = 'id STRING, name STRING'
df1 = spark.createDataFrame(data1, schema1)
display(df1)
data2 = [
            ("3", "rahul"),
            ("4", "jas")
        ]
schema2 = 'id STRING, name STRING'
df2 = spark.createDataFrame(data2, schema2)
display(df2)
```

```python
df1.union(df2).display()
```

```python
data3 = [
    ("minhaz", "5"),
    ("monir", "6")
]
schema3 = 'name STRING, id STRING'
df3 = spark.createDataFrame(data3, schema3)
df3.display()
```

```python
df1.union(df3).display()
```

## Thats where UNION BY NAME comes

```python
df1.unionByName(df3).display()
```

## DATE Functions: CurrentDate(), date_add(), date_sub()

```python
df = df.withColumn("Current_date", current_date())
```

```python
df.limit(5).display()
```

```python
df  = df.withColumn("week_after", date_add("Current_date", 7))
df.limit(5).display()
```

```python
df.withColumn("week_before", date_sub("Current_date", 7)).limit(5).display()
```

```python
df = df.withColumn("week_before", date_add("Current_date", -7)).limit(5)
df.limit(5).display()
```

##DateDiff

```python
df = df.withColumn("date_diff", date_diff("week_after", "week_before"))
df.limit(5).display()
```

```python
df = df.withColumn("week_before_formated", date_format("week_before", "MM-dd-yyyy"))
df.limit(5).display()
```
