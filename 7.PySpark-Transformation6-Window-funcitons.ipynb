# WINDOW FUNCTIONS

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
df = spark.read.format("csv")\
        .option("header", True)\
            .option("inferSchema", True)\
                .load("/Volumes/workspace/default/volume1/BigMart Sales.csv")
df.limit(5).display()
df.describe()
```

### ROW_NUMBER()

```python
df.withColumn('row_number', row_number().over(Window.orderBy('Item_Identifier'))).display()
```

### RANK() and DENSE_RANK()

```python
df\
    .withColumn('ranked', rank()\
        .over(Window\
            .orderBy("Item_Identifier")
        )
    )\
        .withColumn('dense_ranked', dense_rank()\
            .over(Window\
                .orderBy("Item_Identifier")
            )
        ).display()
```

### CUMULATIVE SUM()

```python
df\
    .withColumn("item_type_sum", sum("Item_MRP")\
        .over(Window.partitionBy("Item_Type")))\
            .display()
```

##CUMULATIVE_SUM()

```python
df\
    .withColumn("cumulative_sum", sum("Item_MRP")\
        .over(Window.orderBy("Item_Type").rowsBetween(Window.unboundedPreceding, Window.currentRow)))\
            .select(["Item_Type", "Item_MRP", "cumulative_sum"]).limit(10).display()
```
