## Handling NULLS


```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/workspace/default/volume1/BigMart Sales.csv")
```

```python
df.display()
```

```python
df1 = df.dropna(how='any') #Another option is how = 'all' which will drop rows with all valus are null
df1.display()
```

```python
df.dropna(subset=["Outlet_Size"]).display()
```

```python
df.display()
```

## Fill NULL values

```python
df.fillna("NotAvailable", subset = ["Outlet_Size"]).display()
```

```python
        
df.fillna({"Item_Weight": 0, "Outlet_Size": "NA"}).display()
```

## SPLIT and Indexing

### SPLIT

```python
df.withColumn("Outlet_Type", split(col("Outlet_Type"), " ")).limit(5).display()
```

### INDEXING

```python
df.withColumn("Outlet_Type", split(col("Outlet_Type"), " ")[1]).limit(5).display()
```

### EXPLODES

```python
df2 = df.withColumn("Outlet_Type", split(col("Outlet_Type"), " "))
df2.display()
```

```python
df2.withColumn("Outlet_Type_explodes", explode(col("Outlet_Type"))).display()
```

### ARRAY CONTAINS

```python
df2.withColumn("tier1Exist", array_contains(col("Outlet_Type"), "Type1")).display()
```

### Scenario 1

```python
df.groupBy("Item_Type").sum("Item_MRP").display()
```

%md
### Scenario 2

```python
df.groupBy("Item_Type").agg(sum("Item_MRP")).display()
```

%md
### Scenario 3

```python
df.groupBy("Item_Type", "Outlet_Size").agg(sum("Item_MRP").alias("Total_MRP")).orderBy("Item_Type").display()
```

### Scenario 4

```python
df.groupBy("Item_Type", "Outlet_Size").agg(sum("Item_MRP").alias("Total_MRP"), avg("Item_MRP").alias("Avg_MRP")).orderBy("Item_Type").display()
```
